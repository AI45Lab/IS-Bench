```python
import json
import os
import re
```

```python
key_word = 'boil_water_in_the_microwave__7'

for file in os.listdir('tasks'):
    if key_word in file: 
        src = os.path.join('tasks', file)
        new_file = file.replace('7_', '')
        new_file = new_file.replace('_on_cabinet', '')
        new_file = new_file.replace('add', 'with')
        task_name = new_file.replace('.json', '')
        dst = os.path.join('tasks', new_file)
        with open(src) as f:
            task_config = json.load(f)
            task_config['task_info']['task_name'] = task_name
        # print(src)
        print(dst)
        os.rename(src, dst)
        json_str = json.dumps(task_config, indent=2)
        with open(dst, 'w') as f:
            f.write(json_str)
```
